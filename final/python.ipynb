```python
import duckdb
import altair as alt
```

```python
info = duckdb.sql("""SELECT  *  
                        FROM read_csv_auto('precios-en-surtidor-resolucin-3142016 (2).csv')
                        WHERE indice_tiempo =='2023-11' AND 
                            empresabandera == 'YPF' AND 
                            provincia == 'CAPITAL FEDERAL' AND 
                            producto == 'Nafta (s√∫per) entre 92 y 95 Ron' AND
                            tipoHorario= 'Diurno'
                        ORDER BY precio DESC
                  """).df()


direccion = info['direccion'][0]
precio = info['precio'][0]
print(f"La estacion de servicio mas cara con un precio de {precio} esta en la direccion: {direccion}")

```

```python
alt.Chart(info).mark_bar().encode(
    x=alt.X('precio', bin=True),
    y='count()',
    tooltip=['precio','count()']
)
```
